```python
#for pie chart
labels=['A','B','C','D']
sizes=[20,40,10,30]
plt.figure(figsize=(5,5))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140,colors=['red','green','gold','blue'])
plt.title('pie chart')
plt.show()
```

```python
a=[1,2,3,4]
b=[2,4,6,7]
c=[6,4,3,2]
plt.boxplot([a,b,c],labels=['a','b','c'])
plt.title('boxplot')
plt.show()
```

```python
data=[[1,2,3],
     [4,5,6],
      [7,8,9]]
sns.heatmap(data,annot=True)
plt.title('h')
plt.show()
```
