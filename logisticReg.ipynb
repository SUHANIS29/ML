```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Users\suhan\Downloads\Expanded_Logistic_Regression_Data.csv")
df
```

```python
plt.scatter(df.Age,df.Has_Insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test= train_test_split(df[['Age']],df.Has_Insurance,test_size=0.1)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
