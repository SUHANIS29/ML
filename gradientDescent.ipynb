```python
import numpy as np
def gradient_descent(x,y):
    m_curr=b_curr=0
    iteration=10000
    n=len(x)
    learning_rate=0.08
    
    for i in range(iteration):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
        md=-(2/n)*sum(x*(y-y_predicted))
        bd=-(2/n)*sum(y-y_predicted)
        m_curr=m_curr-learning_rate*md
        b_curr=b_curr-learning_rate*bd
        print("m {},b {},cost {}, iteration {}".format(m_curr,b_curr,cost,i))
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent Function with Visualization
def gradient_descent_visual(x, y):
    m_curr = b_curr = 0  # Initial values of m (slope) and b (intercept)
    iterations = 100  # Number of iterations
    n = len(x)  # Number of data points
    learning_rate = 0.08

    # Store the cost and line parameters for visualization
    costs = []
    lines = []

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1 / n) * sum([val ** 2 for val in (y - y_predicted)])
        md = -(2 / n) * sum(x * (y - y_predicted))  # Gradient w.r.t. m
        bd = -(2 / n) * sum(y - y_predicted)       # Gradient w.r.t. b
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        # Save cost and line parameters for visualization
        costs.append(cost)
        lines.append((m_curr, b_curr))

    return costs, lines

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Perform gradient descent
costs, lines = gradient_descent_visual(x, y)

# Plot the data and the regression lines
plt.figure(figsize=(10, 6))

# Plot original data points
plt.scatter(x, y, color='red', label='Data Points')

# Plot regression lines during iterations
for i in range(0, len(lines), 10):  # Plot every 10th line to reduce clutter
    m, b = lines[i]
    y_pred = m * x + b
    plt.plot(x, y_pred, label=f'Iteration {i}')

plt.title('Gradient Descent - Line Adjustments')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Plot the cost reduction over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(len(costs)), costs, color='blue')
plt.title('Cost Reduction Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.show()
```
