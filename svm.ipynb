```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
iris.target_names
#provide array of target names i.e. the groups in which we hv to classify give data of length n width
```

```python
df[df.target==1] #give data of versicolor grp
```

```python
df['flower_name']=df.target
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker='o')
```

```python
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker='o')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='red', marker='*')
```

```python
plt.xlabel('petal length(cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', marker='o')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='red', marker='*')
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(['target','flower_name'],axis='columns')
X.head()
```

```python
y=df.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
# model=SVC(C=10
model=SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
